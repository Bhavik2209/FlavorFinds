```python
def get_title(soup):
    try:
        title = soup.find('h1',attrs={'class':'biGQs _P egaXP rRtyp'})
        title_value = title.text
        title_string = title_value.strip()
    except AttributeError:
        title_string = ''
    return title_string

def get_address(soup):
    try:
        address = soup.find('div',attrs={'class':'biGQs _P pZUbB hmDzD'}).text
    except AttributeError:
        address=''
    return address

def get_timing(soup):
    try:
        t1 = soup.find_all('span',attrs={'class':'biGQs _P pZUbB egaXP hmDzD'})
        if(len(t1)>1):
            timing = t1[0].text + " and " + t1[1].text
        else:
            timing=''
    except AttributeError:
        timing =''
    return  timing

def get_type(soup):
    try:
        type1 = soup.find_all('div',attrs={'class' : 'biGQs _P pZUbB alXOW oCpZu GzNcM nvOhm UTQMg ZTpaU W hmDzD'})
        if(len(type1)>0):
            type = type1[0].text
        else:
            type=''
    except AttributeError:
        type =''
    return type

def get_type2(soup):
    try:
        type2 = soup.find_all('div',attrs={'class' : 'biGQs _P pZUbB alXOW oCpZu GzNcM nvOhm UTQMg ZTpaU W hmDzD'})
        if(len(type2)>1):
            type = type2[1].text
        else:
            type=''
    except AttributeError:
        type =''
    return type    

def get_description(soup):
    try:
        description1 = soup.find_all('div',attrs={'class' : 'biGQs _P pZUbB alXOW oCpZu GzNcM nvOhm UTQMg ZTpaU W hmDzD'})
        if(len(description1)>2):
            description = description1[2].text
        else:
            description=''
    except AttributeError:
        description =''
    return description

def get_phone(soup):
    try:
        phone1 = soup.find_all('span',attrs={'class' : 'biGQs _P pZUbB hmDzD'})
        if(len(phone1)>0):
            phone = phone1[-1].text
        else:
            phone=''
    except AttributeError:
        phone =''
    return phone

def get_total_ratings(soup):
    try:
        total_rating = soup.find('span',attrs={'class' : 'biGQs _P fiohW oXJmt'}).text
    except AttributeError:
        total_rating = ''
    return total_rating

def get_ratings(soup):
    try:
        rating = soup.find('span',attrs={'class' : 'biGQs _P fiohW uuBRH'}).text
    except AttributeError:
        rating = ''
    return rating
```
